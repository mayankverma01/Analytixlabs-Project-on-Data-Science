```python
#There are 133,600 missing values in the CustomerID column ,we will remove these missing values
O_Retail=O_Retail[pd.notnull(O_Retail['CustomerID'])]
O_Retail.info()
```

```python
O_Retail=O_Retail[O_Retail['Quantity']>0]
O_Retail.Quantity.min()
```

```python
#total Price
O_Retail['TotalPrice']=O_Retail['Quantity']*O_Retail['UnitPrice']
```

```python
#Since recency is calculated for a point in time, and the last invoice date is 2011–12–09,
#we will use 2011–12–10 to calculate recency.
import datetime as dt
print(O_Retail['InvoiceDate'].min())
print(O_Retail['InvoiceDate'].max())
Now=dt.datetime(2011,12,10)
O_Retail['InvoiceDate']=pd.to_datetime(O_Retail['InvoiceDate'])
```

```python
O_Retail.head()
```

```python
rfmTable = O_Retail.groupby('CustomerID').agg({'InvoiceDate': lambda x: (Now - x.max()).days, 'InvoiceNo': lambda x: len(x), 'TotalPrice': lambda x: x.sum()})
rfmTable['InvoiceDate'] = rfmTable['InvoiceDate'].astype(int)
rfmTable.rename(columns={'InvoiceDate': 'Recency', 
                         'InvoiceNo': 'Frequency', 
                         'TotalPrice': 'Monetary'}, inplace=True)
```

```python
rfmTable.head()
```

```python
def R_Quant(x):
    Q1=rfmTable.Recency.quantile(0.25)
    Q2=rfmTable.Recency.quantile(0.50)
    Q3=rfmTable.Recency.quantile(0.75)
    Q4=rfmTable.Recency.quantile(1)
    if x <=Q1:
        return 1
    elif x <= Q2:
        return 2
    elif x <= Q3: 
        return 3
    else:
        return 4
    
def F_Quant(x):
    Q1=rfmTable.Frequency.quantile(0.25)
    Q2=rfmTable.Frequency.quantile(0.50)
    Q3=rfmTable.Frequency.quantile(0.75)
    Q4=rfmTable.Frequency.quantile(1)
    if x <=Q1:
        return 4
    elif x <= Q2:
        return 3
    elif x <= Q3: 
        return 2
    else:
        return 1
    
def M_Quant(x):
    Q1=rfmTable.Monetary.quantile(0.25)
    Q2=rfmTable.Monetary.quantile(0.50)
    Q3=rfmTable.Monetary.quantile(0.75)
    Q4=rfmTable.Monetary.quantile(1)
    if x <=Q1:
        return 4
    elif x <= Q2:
        return 3
    elif x <= Q3: 
        return 2
    else:
        return 1
    
```

```python
rfmTable['R_Quantiles']=rfmTable['Recency'].apply(R_Quant)
rfmTable['M_quantiles']=rfmTable['Monetary'].apply(M_Quant)
rfmTable['F_quantiles']=rfmTable['Frequency'].apply(F_Quant)
```

```python
rfmTable.head()
```

```python
rfmTable['RFMScore']=rfmTable.R_Quantiles.map(str)+rfmTable.M_quantiles.map(str)+rfmTable.F_quantiles.map(str)
```

```python
rfmTable.head()
```

```python
rfmTable=rfmTable.assign(R_seg=np.nan, FM_seg =np.nan)
rfmTable.R_seg[rfmTable['R_Quantiles']==4] = "1.Active"
rfmTable.R_seg[rfmTable['R_Quantiles']==2] = "2.At Risk"
rfmTable.R_seg[rfmTable['R_Quantiles']==3] = "2.At Risk"
rfmTable.R_seg[rfmTable['R_Quantiles']==1] = "3.Churn"

rfmTable.FM_seg[(rfmTable['F_quantiles']==3) & (rfmTable['M_quantiles']==1)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==4) & (rfmTable['M_quantiles']==1)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==2) & (rfmTable['M_quantiles']==2)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==3) & (rfmTable['M_quantiles']==2)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==4) & (rfmTable['M_quantiles']==4)] = "1.Silver"
rfmTable.FM_seg[(rfmTable['F_quantiles']==1) & (rfmTable['M_quantiles']==3)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==2) & (rfmTable['M_quantiles']==3)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==3) & (rfmTable['M_quantiles']==3)] = "1.Silver"
rfmTable.FM_seg[(rfmTable['F_quantiles']==4) & (rfmTable['M_quantiles']==3)] = "1.Silver"
rfmTable.FM_seg[(rfmTable['F_quantiles']==1) & (rfmTable['M_quantiles']==4)] = "2.Gold"
rfmTable.FM_seg[(rfmTable['F_quantiles']==2) & (rfmTable['M_quantiles']==4)] = "1.Silver"
rfmTable.FM_seg[(rfmTable['F_quantiles']==3) & (rfmTable['M_quantiles']==4)] = "1.Silver"
rfmTable.FM_seg[(rfmTable['F_quantiles']==4) & (rfmTable['M_quantiles']==4)] = "0.Standard"

rfmTable.FM_seg.fillna("4.Premium", inplace=True)
```
