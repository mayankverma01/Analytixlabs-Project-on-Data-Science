```python
# Importing the two files LeftR and rightR

Lt=pd.read_csv("C:\\Users\\HP\\Desktop\\New folder (2)\\LeftR.csv")
Rt=pd.read_csv("C:\\Users\\HP\\Desktop\\New folder (2)\\RightR.csv")
```

```python
#Inner join - only the common recrds in both table
pd.merge(left=Lt,
         right=Rt,
         left_on=["ID"],
         right_on=["C.ID"],
         how="inner",
         indicator=True)
```

```python
#Left Join- only common record in left table and common matched records from the right tables.
pd.merge(left=Lt,
         right=Rt,
         left_on=["ID"],
         right_on=["C.ID"],
         how="left",
         indicator=True)
```

```python
#Full Join-  common  records from both the  tables.
pd.merge(left=Lt,
         right=Rt,
         left_on=["ID"],
         right_on=["C.ID"],
         how="outer",
         indicator=True)
```
